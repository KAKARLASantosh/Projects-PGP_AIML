```python
import numpy as np
import nltk
import json
import tensorflow as tf
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

<span style="font-family: Arial; font-weight:bold;font-size:2.5em;color:#00b3e5;"> Part 1

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 1. Import and analyse the data set

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=10000)
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 2. Perform relevant sequence adding on the data

```python
x_train = pad_sequences(x_train, maxlen=20)
x_test =  pad_sequences(x_test, maxlen=20)
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 3. Perform following data analysis

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 5. Design, train, tune and test a sequential model

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(10000, 128, trainable=True, input_length=20))
model.add(tf.keras.layers.LSTM(units=64, dropout=0.5))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=250, verbose=1)
```

```python
model_score = model.evaluate(x_test, y_test, verbose=0)
model_score
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 6. Use the designed model to print the prediction on any one sample

```python
y_pred = np.round(model.predict(x_test))
```

```python
y_pred = y_pred.astype('int64')
```

```python
print(classification_report(y_test, y_pred, target_names=['Positive sentiment', 'Negative sentiment']))
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 2. Retain relevant columns

```python
data.drop('article_link', axis=1,inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
print(data['headline'][0])
print(data['headline'][1])
print(data['headline'][200])
print(data['headline'][2000])
print(data['headline'][20000])
```

```python
data['headline'] = data['headline'].apply(lambda x: x.lower())
data['headline'] = data['headline'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))
data['headline'] = data['headline'].apply(lambda s: s.strip())
```

```python
from string import punctuation
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english') + list(punctuation)
```

```python
data['headline'] = data['headline'].apply(lambda text: ' '.join([words for words in text.split() if words not in stop_words]) )
```

```python
data.head()
```

```python
print(data['headline'][0])
print(data['headline'][1])
print(data['headline'][200])
print(data['headline'][2000])
print(data['headline'][20000])
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 3. Get length of each sentence

```python
len_array = [(len(text)) for text in data['headline']]
```

```python
len_vec = np.array(len_array)
len_vec
```

```python
print('Maximum length of a sentence:', np.max(len_vec))
print('Position of the sentence with maximum length:', np.argmax(len_vec))
```

```python
data['headline'].iloc[np.argmax(len_vec)]
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 4. Define parameters

```python
max_features = 10000
maxlen=50
```

```python
tokenizer = Tokenizer(num_words=max_features,filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',lower=True,split=' ', char_level=False)
tokenizer.fit_on_texts(data['headline'])
```

```python
X = tokenizer.texts_to_sequences(data['headline'])
len_array_2 = [(len(l)) for l in X]
len_vec_2 = np.array(len_array_2)

print('Maximum length of a sequence:', np.max(len_vec))
print('Position of the sequence with maximum length:', np.argmax(len_vec))
```

- Since we retain only first 10,000 repeated words, the sentence with 731 words converted to a sequence of only 81 indices. 
- However, as we are given a 50 dimensional Glove embeddings, we pad the feature list with 50 as the maximum length.

```python
X = pad_sequences(X, maxlen=maxlen)
y = np.asarray(data['is_sarcastic'])

print('Feature list of the sequence with maximum length:',X[np.argmax(len_vec_2)])
print('Label of the sequence with maximum length:',y[np.argmax(len_vec_2)])
```

```python
print(X[0])
print(y[0])
```

```python
print(np.unique(y))
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 5. Get indices for words

```python
word_index = tokenizer.word_index
```

```python
word_index
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 6. Create features and labels

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 7. Get vocabulary size

```python
vocab_size=len(tokenizer.word_index)
print(vocab_size)
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 8. Create a weight matrix using GloVe embeddings

```python
wt_mat = np.empty((vocab_size, 50))

for txt in open('./glove.6B.50D.txt'):
    word = txt.split(" ")[0]
    embd = txt.split(" ")[1:]
    if word in word_index.keys():
        wt_mat[word_index[word]] = np.asarray(embd, dtype='float32')
```

```python
wt_mat
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 9. Define and compile a Bidirectional LSTM model

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size, 128, trainable=True, input_length=50))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=64, dropout=0.5)))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

<span style="font-family: Arial; font-weight:bold;font-size:2em;color:#00b3e5;"> 10. Fit the model and check the validation accuracy

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=250, verbose=1)
```

- The validation accuracy of the model is 79.32%

```python
y_pred = np.round(model.predict(X_test))
```

```python
y_pred = y_pred.astype('int64')
```

```python
print(classification_report(y_test, y_pred,target_names=['Positive sentiment', 'Negative sentiment']))
```
